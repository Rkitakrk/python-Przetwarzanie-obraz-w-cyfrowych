```python
def binaryzacja(img, level_dwon, level_up, status):
    if status:
        for i in range(img.shape[0]):
            for j in range(img.shape[1]):
                if img[i,j] >= level_down and img[i,j] <= level_up:
                    img[i,j] = 255
                else:
                    img[i,j] = 0

        return img
    else:
        for i in range(img.shape[0]):
            for j in range(img.shape[1]):
                if img[i,j] >= level_down and img[i,j] <= level_up:
                    img[i,j] = 0
                else:
                    img[i,j] = 255

        return img      

```

```python
def wyswietl(img, img1, img2, size=(10,10)):
    import matplotlib.pyplot as plt
    
    f, (ax1, ax2, ax3) = plt.subplots(1,3, figsize=size)
    
    ax1.imshow(img, cmap='gray')
    ax1.axis('off')
    ax2.imshow(img1, cmap='gray')
    ax2.axis('off')
    ax3.imshow(img2, cmap='gray')
    ax3.axis('off')
    plt.show()
```

```python
from skimage import data
import matplotlib.pyplot as plt

img = data.coins()
img1 = data.coins()
img2 = data.coins()

level_up = 200
level_down = 150

plt.imshow(img, cmap="gray")
plt.axis('off')
plt.show()
img1 = binaryzacja(img1, level_down, level_up, True)
plt.imshow(img1, cmap="gray")
plt.axis('off')
plt.show()
img2 = binaryzacja(img2, level_down, level_up, False)

plt.imshow(img2, cmap="gray")
plt.axis('off')
plt.show()

wyswietl(img, img1, img2)
```
